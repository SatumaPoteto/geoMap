### たつの市のマップを作成してみる

```python
import pandas as pd 
import geopandas as gpd
import matplotlib.pyplot as plt
import glob
```

```python
# 国勢調査の境界データ(shpファイル)を抽出
shp_files = glob.glob("*.shp")
print(shp_files)
```

```python
# shpファイルをdataflameに変換していく
shpdf = []
for f in shp_files:
    shp_df = gpd.read_file(f)
    shpdf.append(shp_df)
```

```python
# 試しにプロットしてみる
display(shpdf[1].head());shpdf[1].info();shpdf[1].shape
```

```python
# それぞれplotしてみる
tatuno = shpdf[1]
hyogo = shpdf[0]
# たつの城下町のみを抽出
tatunojouka = shpdf[1][shpdf[1]["MOJI"].str.match('龍野町*')]
```

```python
# 龍野町の名前をリスト化
print(list(tatuno[tatuno.S_NAME.str.match("龍野*.")].S_NAME))
```

```python
# 龍野市全体のプロット
fig= plt.figure(figsize=(6,10))
ax = fig.add_subplot(111)
tatuno.plot(ax=ax,cmap="summer",column="MOJI")
ax.axis(False)
plt.title("tatuno",fontsize=25)
plt.tight_layout()
plt.savefig("tatuno.png")
```

```python
# 龍野町のみをプロット
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111)
tatunojouka.plot(ax=ax,cmap="summer",column="MOJI")
ax.axis(False)
plt.title("tatuno mati",fontsize=25)
plt.tight_layout()
plt.savefig("tatuno_mati.png")
plt.show()
```
